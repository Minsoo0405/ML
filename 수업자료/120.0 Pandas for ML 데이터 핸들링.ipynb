```python
titanic_df['gender_person'] = titanic_df['Sex'].apply(lambda x: x + '_person')
titanic_df
```

```python
# 타이타닉 'Age'열의 정보 중 18세 이상 22세 이하이면 'Young'의 값을 가지고, 아니면 'Not Young'
# 데이터를 가지는 Age_Cat'열을 생성하라
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Young' if 18 <= x <= 22 else 'Not Young')
titanic_df
```

```python
def get_age_cat(x):
    if x <= 5:
        cat = 'Baby'
    elif x <= 12:
        cat = 'Child'
    elif x <= 18:
        cat = 'Teenager'
    elif x <= 25:
        cat = 'Student'
    elif x <= 35:
        cat = 'Young Adult'
    elif x <= 60:
        cat = 'Adult'
    else:
        cat = 'Elderly'
    return cat
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(get_age_cat)
titanic_df
```

### 데이터 전처리

#### One-hot encoding, Scaling

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자렌지', '컴퓨터', '선풍기', '선풍기', '믹서', '믹서']

encoder = LabelEncoder()
encoder.fit(items)
```

```python
# ASII코드에서 UNI코드로 변환되면 영어, 한글 각각 abc, 가나다순으로 정렬
labels = encoder.transform(items)
labels
```

```python
# 라벨 인코딩 규칙
encoder.classes_
```

```python
# 원본 값 확인
encoder.inverse_transform(labels)
```

```python
# 먼저 라벨인코딩으로 변환 후 원핫 인코딩

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items = ['TV', '냉장고', '전자렌지', '컴퓨터', '선풍기', '선풍기', '믹서', '믹서']

# 라벨 인코딩
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)

# 원핫 인코딩으로 변환하기 전에 2차원 배열로 구조 변경
# 구조의 모양을 튜플로 넣어야함.
labels = labels.reshape((-1, 1))

# 원핫 인코딩
oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)
oh_labels.toarray()
```

```python
# 원본 값 확인
oh_encoder.inverse_transform(oh_labels)
```

데이터 스케일링

    1. 피처 스케일링(Feature Scaling)
        : 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업
    2. 표준화(Standardization)
        : 평균 0, 분산 1인 가우시안 정규 분포를 가진 값으로 변환
    3. 정규화(Nomalization)
        : 서로 다른 피처의 크기를 통일하기 위해 크기를 변환

    4. 사이킷런 피처 스케일링
        : 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업
        - StandardScaler()
        - MiinMaxCsaler()

```python
from sklearn.datasets import load_iris

import pandas as pd

iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df
```

```python
df = pd.DataFrame(data=iris.data)
df.describe()
```

```python
# StandardScaler()
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
np.set_printoptions(precision=6, suppress=True)
iris_scaled_df = pd.DataFrame(iris_scaled)
pd.options.display.float_format = '{:.5f}'.format
iris_scaled_df.describe()
```

```python
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(iris.data)
scaler.transform(iris.data)
```
