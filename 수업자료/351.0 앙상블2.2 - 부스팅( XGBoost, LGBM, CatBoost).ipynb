```python
def get_eval_score(y_test, y_pred, y_pred_proba):
    confusion = confusion_matrix(y_test, y_pred)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    
    print("-"* 100)
    print("오차행렬:")
    print(confusion)
    print("정확도: {:.4f} 정밀도: {:.4f} 재현율: {:.4f} F1:{:.4f} AUC: {:.4f}".\
          format(acc, prec, recall, f1, roc_auc ))
    return
```

```python
### 1. 조기종료 옵션을 사용하지 않는 경우
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=400, learning_rate=0.1,
              max_depth=3, random_state=156)

xgb.fit(X_train,y_train, eval_metric='error')
```

```python
get_eval_score(y_train, xgb.predict(X_train), xgb.predict_proba(X_train)[:,1])

get_eval_score(y_test, xgb.predict(X_test), xgb.predict_proba(X_test)[:,1])
```

```python
## 2. 조기 종료 옵션을 사용한 경우
### early stop 옵션(eval_set, eval_metric, early_stopping_rouns, verbose, callbacks, ...)
xgb.fit?
```

```python

xgb2 = XGBClassifier(n_estimators=400, learning_rate=0.1,
              max_depth=3)

xgb2.fit(X_train, y_train, early_stopping_rounds=100,
         eval_metric='logloss', eval_set=[(X_test, y_test)],
         verbose=True)

```

```python
xgb2.best_iteration
```

```python
* 하이퍼파라미터 튜닝( 조기종료 옵션 사용)
```

```python
# 조기종료 모델에 대한 평가(early_stopping_rounds:100)
get_eval_score(y_train, xgb2.predict(X_train), xgb2.predict_proba(X_train)[:,1])

get_eval_score(y_test, xgb2.predict(X_test), xgb2.predict_proba(X_test)[:,1])
```

```python
# 조기종료 모델 구축 및 평가(early_stopping_rounds=10)
xgb3 = XGBClassifier(n_estimator=400, learning_rate=0.1,
              max_depth=3)

xgb3.fit(X_train, y_train, early_stopping_rounds=10,
         eval_metric='logloss', eval_set=[(X_test, y_test)],
         verbose=True)
```

```python
# 조기종료 모델에 대한 평가(early_stopping_rounds:10)
get_eval_score(y_train, xgb3.predict(X_train), xgb3.predict_proba(X_train)[:,1])

get_eval_score(y_test, xgb3.predict(X_test), xgb3.predict_proba(X_test)[:,1])
```
