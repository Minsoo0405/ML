# 1. 군집(Clustering)


- 군집의 개요 
    - 개념
        - 데이터 클러스터 세트의 관측값이 가지고 있는 여려 속성을 분석ㅎ여 소러 비슷한 속성을 갖는 관측값끼리 같은(군집)으로 묶는 알고리즘
        - 다른 클러스터 간에는 서로 와전하게 구분되는 특징을 갖기 때문에 어느 클러스터에도 속하지 못하는 관측값(특이 데이터, 이상값, 중복값)을 찾는데도 유용
    - 목적
        - 데이터로부터 유의미한 그룹들을 분리하는 것
    - 분류 알고리즘과의 차이
        - 공통점: 데이터를 비슷한 집단으로 묶는다
        - 차이점: 분류는 소속 집단의 정보를 미리 알고 있는 상태에서 비슷한 집단으로 묶는 것이며, 군집화란 소속집단의 정보가 없는 상태에서 비슷한 집단으로 묶는 방식(비지도학습)
    - 대표 알고리즘
        - K-Means
        - DBSCAN(Density-Based Spatial clustering of Applications with Noise)
        - 계층적 집단: 덴드로그램


- 군집의 기능
    - 변수와 레코드가 아주 큰상황인 경우 탐색적 데이터 분석의 연장으로 활용ㅇ
    - 콜드 스타트 문제(아무런 정보가 없는 경우)에서 유용한 방법
    - 모델을 훈련시킬 수 있는 응답 데이터를 초기에 가지고 있지 않은 경우, 패턴이 비슷한 데이터들을 분류하여 학습과정을 빨리 시작할 수 있도록 함
    - 군집의 결과로 얻은 그룹을 직접 사용할 수도 있고, 예측을 위한 회귀나 분류 모델의 입력 피쳐 또는 타겟 변수로 사용할 수도 있음.

- 군집 분석의 활용
    - 소비자 행동 패턴 그룹화 - 신용카드 부정 사용 탐지, 구매 패턴 분석, 택배 배송 경로 개선
    - 마케팅: 시장 세분화, 고객 세분화
    - 금융 - 포트폴리오 구축 세분화, 투자정보 데이터, 재무 상황, 산업 및 시장의 자본 총액과 같은 특성치를 이용해 특징이 유사한 종목을 군집화, 또는 서로 다른 군집으로부터 주식을 선택하여 균형있는 포트폴리오 구축
    - 온라인 쇼핑몰 - 유사 상품을 탐색하거나 구매하는 고객들을 묶어 각 그룹이 보다 관심을 가진만한 쇼핑정보를 제공하여 구매를 유도하거나 고객의 이탈을 방지
    - 대형마트 - 고객 장바구니 목록을 분석하여 서로 같이 구매되는 제품들을 상품으로 기획하거나 진열에 참고
    - Mendeleev 원소 주기율표 - 군집 개념을 성공적으로 설명한 예
    - 생명정보학, 의학, 사회과학 등 다양한 분야에서 활용
    - 어떤 지역의 식생 유형을 파악하려는 경우
    - 문서, 음악, 영화를 여러 주제의 그룹으로 모으는 경우
    - 구매이력의 공통 부분을 기반으로 관심사가 비슷한 고객을 찾는 것

# 2. K-Means(K-평균) 알고리즘

- 개념
    - 원리가 단순하고 직관적이고 성능이 좋은 군집 알고리즘
    - 데이터 크기가 커져도 손쉽게 사용 가능
    - 분할적 군집 분석(군집의 수를 미리 결정한 뒤 군집화 수행)
    - 각 관측자는 K개 군집 중 하나에 무조건 속하며, 2개 이상의 군집에 속하지 않음
    - 데이터 간 유사성을 측정하는 기준: 각 군집(클러스터)의 중심(센트로이드)까지의 거리
    - 데이터 정규화 필요성: 정규화가 없는 경우 스케일이 가장 큰 변수가 군집 과정을 독점하게 됨
    - 몇 개의 군집(K)으로 데이터를 구분할 것인가에 따라 모형의 성능이 달라짐
    - 주어진 k에 선택한 분할 기준을 최적화할 수 있는 k개의 군집을 찾음
        - 반복적으로 수많은 발을 수행 검토함
        - 각 군집은 중심에 의해 표현됨
        - 객체들이 어느 증심점에 가까운가에 따라 군집을 결정함 

- 작동 방식
    - 현재 가지고 있는 데이터 중 임의의 두 점을 선택
    - 이 두 점과 가장 가까운 점을 그룹으로 묶음
    - 각 그룹의 중심점을 구함
    - 새로운 중심점을 기준을 다시 그룹으로 묶음
    - 이 과정을 반복하여 군집 배정이 같게 되면 종료

- 장점
    - 일반적인 군집화헤서 가장 많이 사용되는 알고리즘
    - 알고리즘이 쉽고 간결함
    - 원형 클러스터를 구분하는데 뛰어남
- 단점
    - 거리 기반 알고리즘으로 속성의 개수가 매우 많을경우 군집화 정확도가 떨어짐
    - 사전에 클러스터 개수 k를 지정해야함
    - 반복을 수행하는데 반복횟수가 많을 경우 수행시간이 매우 느려짐
    - 몇 개의 군집을 선택해야할지 가이드가 어려움
    - K-means는 군집들이 다양성을 가질 때 효율적이지 못함
        - 군집의 크기
        - 군집의 밀도
        - 비원형 형태의 군집

- 최적의 K(클러스터) 찾기
    - 군집개수 K를 결정하는 방법
        - 군집분석이 정답이 없는 상태에서 데이터를 분석하는 비지도학습이다보니 최적의 군집의 개수k를 결정하고 군집 결과를 평가흔ㄴ 것이 다분히 주관적인 편
        - 사전에 산점도 등을 통해 데이터 특성 등을 확인하여 데이터가 뭉쳐 있는 그룹을 K로 선택하는 방법도 있음
        - 변수의 숫자가 많은 경우 차원축소 등의 방법을 이용해 변수의 수를 줄여서 특성을 파악하기도 함
        
        - 1) 계층적 군집분석의 덴드로그램 시각화를 이용한 군집의 개수 결정
        - 2) 팔꿈치 방법을 이요한 군집의 개수 결정(The Elbow Method)
        - 3) 실루엣 방법을 이용한 군집의 개수 결정(The Silhouette Method)

    - 엘보우 방식(Elbow)
        - ㅇKMeans에서 제공하는 이너셔를 활용
        - 이너셔: 군집의 중심 ~ 개별 샘플 간의 거리의 제곱합
        - 군집에 속한 샘플이 얼마나 가깝게 모여 있는지 나타내는 값
        - 군집의 수가 늘어나면 개별 군집의 크기는 줄어드므로 이너셔도 줄어든다. 엘보우 방법은 군집의 개수를 늘려가면서 이너셔의 변환을 관찰
        - 그래프로 그리면 감소의 속도가 꺾이는 지점이 있는데 이 지점부터는 군집의 개수를 늘려도 군집의 밀집도가 개선되지 않음


# 3. 군집평가

- 실루엣 분석(Silhouette Analysis)
    - 각 군집 간의 거리가 얼만 ㅏ효율적으로 분리되었는지 나타냄.
    - 클러스터 내 데이터들이 얼마나 조밀하게 모여있는지 측정하는 도구
    - 다른 군집과 거리는 떨어져 있고 동일 군집끼리의 데이터는 가깝게 잘 뭉처여있어야 함
    - 군집화가 잘 이루어질수록 개별 군집은 비숫한 정도의 여유 공간을 가지고 떨어져 있음
- 실루엣 계수(Silhoutte Coefficient)
    - 정의
        - 개별 데이터가 가지는 군집화 지표
        - 해당 데이터가 같은 군집 내의 데이터와 얼마나 가깝게 군집화되어있고 다른 군집에 있는 데이터와는 얼마나 멀리 분리되어 있는지 나타내는 지표
        - 특징
            - (-1) < 실루엣 계수 > 1
            - 1에 가까워질수록 근처의 군집과 멀리 떨어져있다
            - 0에 가까울수록 근ㄴ처의 군집과 가까이 있다
            - (-)실루엣 계수: 다른 군집에 데이터 포인트가 할당되었다
- 실루엣 메소드
    - 군집화 결과 판단 기준
        - 전체 실루엣 계수의 평균값이 0 ~ 1 사이의 값을 가지며, 1에 가까울 수록 좋다
        - 개별 군집의 평균값의 차이가 작아야 한다.
        - 개별 군집의 실루엣 계수 평균값이 전체 실루엣 계수의 평균값에서 크게 벗어나지 않는 것이 중요하다
        - 만약 전체 실루엣 계수의 평균값은 높지만, 특정 군집의 실루엣 계수 평균값만 유난히 높고 다른 군집들의 실루엣 계수 평균값은 낮으면 좋은 군집이라고 할수 없다.

# 4. DBSCAN

- 개념
    - Density-Based Spatial Clustering of Applications with Noise
    - 밀도 기반 군집화의 대표적 알고리즘(특정 공간 내의 데이터 밀도차이를 기반으로 함)
    - K-Means와 같이 원형 클러스터를 가정하지 않음
- 장점
    - 간단하고 직관적인 알고리즘
    - 데이터의 분포가 기하학적이 복잡한 데이텃셋에도 효과적인 군집화 가능
    사전에 클러스터의 개수를 지정하지않고, 분석 결과로 클러스터의 개수가 정해짐
- 공간 밀집도를 기준으로 클러스터 구분
    - 핵심(코어)포인트
    - 이웃 포인트
    - 경계 포인트
    - Noise(Outlier)
- 파라미터
    - 입실론(epsilon) 주변 영역 - 핵심포인트로부터의 반경
    - 최소 데이터 개수(min points): 입실론 주변 영역에 포함되는 타 데이터의 개수

# 5. 덴드로그램

- 개념
    - 계층적 클러스터링은 각 레코드 자체를 개별 군집으로 설정하여 시작하고, 가장 가까운 클러스터를 결합해나가는 작업을 반복함(상향식, 병합 군집)
    - 특이점이나 비정상적인 그룹이나 레코드를 발견하는데 민감함
    - 직관적인 시각화가 가능(덴드로(나무), 그램(그림)) 해석이 수월하다
    - 상대적으로 데이터 크기가 작은 문제에 주로 적용되며, 수백만개의 대규모 레코드에는 적용할 수 없다.(유연성, 비용)
- 측정 지표
    - 레코드간의 거리, 군집간의 유사도
    - 트리의 잎: 각각의 표본(레코드)
    - 트리의 가지 길이: 해당 클러스터간의 차이 정도
